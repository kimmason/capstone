```python
import pandas as pd
import re
from soynlp.normalizer import normalize
import unicodedata

# 이모티콘 매핑
EMOTICON_MAP = {
    "❤️": "사랑해요", "🧡": "사랑해요", "💛": "사랑해요", "💚": "사랑해요",
    "💙": "사랑해요", "💞": "사랑해요", "💓": "사랑해요", "💜": "사랑해요",
    "❣️": "사랑해요", "💕": "사랑해요", "💘": "사랑해요", "💗": "사랑해요",
    "💝": "사랑해요", "💟": "사랑해요", "😻": "사랑해요", "💔": "싫어해요",
    "👍": "최고에요", "👎": "최악이에요", "🙌": "만세", "😘": "사랑해요",
    "😍": "사랑해요", "😃": "좋아요", "😄": "좋아요", "😁": "좋아요",
    "😆": "좋아요", "☺️": "좋아요", "😊": "좋아요", "😚": "좋아요",
    "🤗": "좋아요", "😭": "슬퍼요", "😢": "슬퍼요", "😤": "삐졌어요",
    "😠": "화났어요", "😡": "화났어요", "🤬": "화났어요", "😳": "잘 모르겠어요",
    "🤔": "고민해 볼게요", "^^": "좋아요", "♡": "사랑해요", "♥": "사랑해요"
}

# 이모티콘 처리 함수
def replace_emoticons(text):
    """
    Replace emoticons and special symbols in text with mapped expressions.
    Handles non-string types by converting them to strings.
    """
    if not isinstance(text, str):  # 문자열이 아닌 경우 처리
        return str(text)  # 문자열로 변환
    for emoticon, replacement in EMOTICON_MAP.items():
        text = text.replace(emoticon, replacement)
    return text

# EMOTICON_MAP에서 모든 이모티콘 가져오기
allowed_emoticons = re.escape(''.join(EMOTICON_MAP.keys()))

# 정규식 수정: 이모티콘을 제외한 특수문자만 삭제
def clean_sentence(sentence):
    return re.sub(fr'(?![{allowed_emoticons}])[^a-zA-Z0-9가-힣\s!?.,]', '', sentence)

# 텍스트 전처리 함수
def preprocess_text(text):
    # 자모 분리 해결
    text = unicodedata.normalize('NFC', text)
    # 중복 문자 정리
    text = normalize(text)
    return text

# 데이터프레임 가정: final_df
# final_df['Sentence'] 컬럼 전처리
final_df['Sentence'] = final_df['Sentence'].apply(replace_emoticons)  # 이모티콘 매핑
final_df['Sentence'] = final_df['Sentence'].apply(clean_sentence)  # 특수문자 처리
final_df['Sentence'] = final_df['Sentence'].apply(preprocess_text)  # 중복 문자 정리

# 텍스트 길이에 따라 필터링
final_df = final_df[
    final_df['Sentence'].apply(len).between(5, 300)
]

# 길이 분포 확인
print("텍스트 길이 필터링 완료!")
print("필터링 후 데이터 크기:", len(final_df))

# 결과 저장
output_path = "/content/drive/MyDrive/캡스톤/final_dataset.csv"  # 저장 경로
final_df.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f"Dataset saved as CSV at: {output_path}")
```

```python
# KoBERT 모델 및 토크나이저 로드
from transformers import BertModel
bertmodel = BertModel.from_pretrained('skt/kobert-base-v1')
tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')
```

```python
import torch

class BERTDataset(torch.utils.data.Dataset):
    def __init__(self, dataset, tokenizer, max_len):
        """
        BERTDataset 초기화

        Args:
            dataset (pd.DataFrame): Sentence와 Emotion 컬럼이 포함된 데이터프레임
            tokenizer: BERT 모델에 맞는 토크나이저
            max_len (int): 최대 시퀀스 길이
        """
        self.sentences = dataset['Sentence'].tolist()
        self.labels = dataset['Emotion'].tolist()
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        """
        주어진 인덱스에서 데이터 반환

        Args:
            idx (int): 데이터 인덱스

        Returns:
            dict: 'input_ids', 'attention_mask', 'labels' 키로 구성된 텐서 데이터
        """
        try:
            # 토큰화 및 인코딩
            encoded = self.tokenizer(
                self.sentences[idx],
                max_length=self.max_len,
                padding='max_length',
                truncation=True,
                return_tensors="pt"
            )
            # 데이터 반환
            return {
                'input_ids': encoded['input_ids'].squeeze(0),
                'attention_mask': encoded['attention_mask'].squeeze(0),
                'labels': torch.tensor(self.labels[idx], dtype=torch.long),
            }
        except Exception as e:
            print(f"Error processing index {idx}: {e}")
            raise
```

```python
import torch.nn as nn
from transformers import AdamW
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

# KoBERT 분류기 정의
class BERTClassifier(nn.Module):
    def __init__(self, bert, dr_rate=0.6):  # Dropout 확률 증가
        super(BERTClassifier, self).__init__()
        self.bert = bert
        self.dropout = nn.Dropout(p=dr_rate)
        self.classifier = nn.Linear(self.bert.config.hidden_size, 3)  # 3 classes

    def forward(self, input_ids, attention_mask):
        output = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = output.pooler_output
        dropped_output = self.dropout(pooled_output)
        return self.classifier(dropped_output)

# Custom Loss: Focal Loss
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        BCE_loss = nn.CrossEntropyLoss()(inputs, targets)
        pt = torch.exp(-BCE_loss)
        F_loss = self.alpha * (1 - pt) ** self.gamma * BCE_loss
        return F_loss
```

```python
# 학습 정확도 계산 함수
def calc_accuracy(X, Y):
    max_vals, max_indices = torch.max(X, 1)
    return (max_indices == Y).sum().data.cpu().numpy() / max_indices.size(0)
```

```python
import os
from itertools import product
import torch
from sklearn.metrics import f1_score
from torch.nn.utils import clip_grad_norm_
from transformers import AdamW
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from tqdm import tqdm

# 하이퍼파라미터 조합
hyperparameter_combinations = [
    {'max_len': 128, 'batch_size': 32, 'num_epochs': 10, 'warmup_ratio': 0.1, 'learning_rate': 3e-5, 'weight_decay': 0.01},
    {'max_len': 256, 'batch_size': 16, 'num_epochs': 20, 'warmup_ratio': 0.2, 'learning_rate': 2e-5, 'weight_decay': 0.01},
    {'max_len': 128, 'batch_size': 64, 'num_epochs': 5, 'warmup_ratio': 0.1, 'learning_rate': 5e-5, 'weight_decay': 0.01}
]

# 결과 저장
results = []
optimal_model_path = "best_overall_kobert_model.pt"
best_overall_acc = 0.0  # 전체 조합 중 최고의 Validation Accuracy
early_stop_patience = 5  # Early stopping patience 설정

# 학습 루프
for idx, params in enumerate(hyperparameter_combinations):
    print(f"\n[Hyperparameter Set {idx+1}] Training...")

    # 데이터셋 재생성
    train_dataset = BERTDataset(train_data, tokenizer, params['max_len'])
    val_dataset = BERTDataset(val_data, tokenizer, params['max_len'])
    test_dataset = BERTDataset(test_data, tokenizer, params['max_len'])

    train_dataloader = DataLoader(train_dataset, batch_size=params['batch_size'], shuffle=True)
    val_dataloader = DataLoader(val_dataset, batch_size=params['batch_size'], shuffle=False)
    test_dataloader = DataLoader(test_dataset, batch_size=params['batch_size'], shuffle=False)

    # 모델 및 옵티마이저 설정
    model = BERTClassifier(bertmodel, dr_rate=0.5).to('cuda')
    optimizer = AdamW(model.parameters(), lr=params['learning_rate'], weight_decay=params['weight_decay'])
    scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=5, T_mult=1)

    # 학습
    best_val_acc = 0.0
    early_stop_counter = 0
    for epoch in range(params['num_epochs']):
        model.train()
        train_loss, train_acc = 0.0, 0.0
        for batch in tqdm(train_dataloader, desc=f"Epoch {epoch+1} Training", leave=False):
            optimizer.zero_grad()
            input_ids = batch['input_ids'].to('cuda')
            attention_mask = batch['attention_mask'].to('cuda')
            labels = batch['labels'].to('cuda')
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            loss = loss_fn(outputs, labels)
            loss.backward()
            clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            train_loss += loss.item()
            train_acc += calc_accuracy(outputs, labels)
        train_loss /= len(train_dataloader)
        train_acc /= len(train_dataloader)

        # Validation
        model.eval()
        val_loss, val_acc = 0.0, 0.0
        all_preds, all_labels = [], []
        with torch.no_grad():
            for batch in tqdm(val_dataloader, desc=f"Epoch {epoch+1} Validation", leave=False):
                input_ids = batch['input_ids'].to('cuda')
                attention_mask = batch['attention_mask'].to('cuda')
                labels = batch['labels'].to('cuda')
                outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                loss = loss_fn(outputs, labels)
                val_loss += loss.item()
                val_acc += calc_accuracy(outputs, labels)
                all_preds.extend(torch.argmax(outputs, axis=1).cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
        val_loss /= len(val_dataloader)
        val_acc /= len(val_dataloader)
        val_f1 = f1_score(all_labels, all_preds, average='weighted')

        print(f"[Epoch {epoch+1}/{params['num_epochs']}] Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Val F1: {val_f1:.4f}")

        # Best Model 저장
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            early_stop_counter = 0
            model_path = f"best_model_set_{idx+1}_epoch_{epoch+1}.pt"
            torch.save(model.state_dict(), model_path)
            print(f"New Best Validation Accuracy for Set {idx+1}: {best_val_acc:.4f} (Model saved at {model_path})")
        else:
            early_stop_counter += 1
            print(f"Early Stop Counter: {early_stop_counter}/{early_stop_patience}")

        if early_stop_counter >= early_stop_patience:
            print("Early stopping triggered.")
            break

    # 테스트 평가
    model.eval()
    test_loss, test_acc = 0.0, 0.0
    with torch.no_grad():
        for batch in test_dataloader:
            input_ids = batch['input_ids'].to('cuda')
            attention_mask = batch['attention_mask'].to('cuda')
            labels = batch['labels'].to('cuda')
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            loss = loss_fn(outputs, labels)
            test_loss += loss.item()
            test_acc += calc_accuracy(outputs, labels)
    test_loss /= len(test_dataloader)
    test_acc /= len(test_dataloader)

    # 전체 결과 저장
    results.append({
        'params': params,
        'val_acc': best_val_acc,
        'test_acc': test_acc
    })

    # 전체 최적 모델 업데이트
    if best_val_acc > best_overall_acc:
        best_overall_acc = best_val_acc
        torch.save(model.state_dict(), optimal_model_path)
        print(f"New Best Overall Model Saved with Validation Accuracy: {best_overall_acc:.4f} at {optimal_model_path}")

# 결과 비교
for idx, result in enumerate(results):
    print(f"\n[Hyperparameter Set {idx+1}] Results:")
    print(f"Params: {result['params']}")
    print(f"Best Validation Accuracy: {result['val_acc']:.4f}")
    print(f"Test Accuracy: {result['test_acc']:.4f}")

print(f"\nBest Overall Validation Accuracy: {best_overall_acc:.4f} (Model saved at {optimal_model_path})")
```

```python
from torch.utils.data import DataLoader

# Train, Validation, Test 데이터셋 준비
train_dataset = BERTDataset(train_data, tokenizer, max_len)
val_dataset = BERTDataset(val_data, tokenizer, max_len)
test_dataset = BERTDataset(test_data, tokenizer, max_len)

# 데이터 로더 준비
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# 데이터 로더 확인
print(f"Train DataLoader size: {len(train_dataloader)}")
print(f"Validation DataLoader size: {len(val_dataloader)}")
print(f"Test DataLoader size: {len(test_dataloader)}")
```

```python
import torch.nn as nn
from transformers import AdamW
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

# KoBERT 분류기 정의
class BERTClassifier(nn.Module):
    def __init__(self, bert, dr_rate=0.6):  # Dropout 확률 증가
        super(BERTClassifier, self).__init__()
        self.bert = bert
        self.dropout = nn.Dropout(p=dr_rate)
        self.classifier = nn.Linear(self.bert.config.hidden_size, 3)  # 3 classes

    def forward(self, input_ids, attention_mask):
        output = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = output.pooler_output
        dropped_output = self.dropout(pooled_output)
        return self.classifier(dropped_output)

# Custom Loss: Focal Loss
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        BCE_loss = nn.CrossEntropyLoss()(inputs, targets)
        pt = torch.exp(-BCE_loss)
        F_loss = self.alpha * (1 - pt) ** self.gamma * BCE_loss
        return F_loss


# 모델 초기화
model = BERTClassifier(bertmodel, dr_rate=0.5).to('cuda')

# 손실 함수 설정 (기존 CrossEntropyLoss를 FocalLoss로 변경)
loss_fn = FocalLoss(alpha=1, gamma=2)

# 옵티마이저 정의
optimizer = AdamW(model.parameters(), lr=learning_rate, weight_decay=weight_decay)

# 스케줄러: CosineAnnealingWarmRestarts
scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=5, T_mult=1)

# 출력 확인
print(f"Model initialized: Dropout Rate={model.dropout.p}, Learning Rate={learning_rate}")
```

```python
from torch.nn.utils import clip_grad_norm_
from sklearn.metrics import classification_report

# 학습 정확도 계산 함수
def calc_accuracy(X, Y):
    max_vals, max_indices = torch.max(X, 1)
    return (max_indices == Y).sum().data.cpu().numpy() / max_indices.size(0)

# 학습 및 평가 루프
best_val_acc = 0.0  # 초기값: 가장 높은 Validation Accuracy 저장
early_stop_counter = 0

for epoch in range(num_epochs):
    print(f"\n[Epoch {epoch + 1}/{num_epochs}] Start Training...")
    model.train()
    train_loss, train_acc = 0.0, 0.0
    for batch in tqdm(train_dataloader, desc=f"Epoch {epoch + 1} Training", leave=False):
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to('cuda')
        attention_mask = batch['attention_mask'].to('cuda')
        labels = batch['labels'].to('cuda')

        # Forward pass
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        loss = loss_fn(outputs, labels)

        # Backward pass
        loss.backward()

        # Gradient clipping
        clip_grad_norm_(model.parameters(), max_norm=1.0)  # Gradient clipping 추가

        # Optimizer step
        optimizer.step()
        scheduler.step()

        train_loss += loss.item()
        train_acc += calc_accuracy(outputs, labels)

        # Memory Optimization
        del outputs, input_ids, attention_mask, labels
        torch.cuda.empty_cache()

    train_loss /= len(train_dataloader)
    train_acc /= len(train_dataloader)
    print(f"[Epoch {epoch + 1}] Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")
    print(f"Learning Rate: {scheduler.get_last_lr()[0]:.6f}")  # Learning Rate 로그 출력

    # Validation
    print(f"[Epoch {epoch + 1}] Start Validation...")
    model.eval()
    val_loss, val_acc = 0.0, 0.0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch in tqdm(val_dataloader, desc=f"Epoch {epoch + 1} Validation", leave=False):
            input_ids = batch['input_ids'].to('cuda')
            attention_mask = batch['attention_mask'].to('cuda')
            labels = batch['labels'].to('cuda')

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            loss = loss_fn(outputs, labels)
            val_loss += loss.item()
            val_acc += calc_accuracy(outputs, labels)

            all_preds.extend(torch.argmax(outputs, axis=1).cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            # Memory Optimization
            del outputs, input_ids, attention_mask, labels
            torch.cuda.empty_cache()

    val_loss /= len(val_dataloader)
    val_acc /= len(val_dataloader)
    val_f1 = f1_score(all_labels, all_preds, average='weighted')
    print(f"[Epoch {epoch + 1}] Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, Val F1: {val_f1:.4f}")

    # Early Stopping & Save Best Model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        early_stop_counter = 0
        best_model_path = f"best_kobert_model_epoch_{epoch + 1}.pt"
        torch.save(model.state_dict(), best_model_path)
        print(f"New Best Validation Accuracy: {best_val_acc:.4f} (Model saved at {best_model_path})")
    else:
        early_stop_counter += 1
        print(f"No improvement for {early_stop_counter} epoch(s)")

    if early_stop_counter >= early_stop_patience:
        print("Early stopping triggered.")
        break

# Test Evaluation
print("\n[Test Evaluation] Start...")
model.eval()
test_loss, test_acc = 0.0, 0.0
all_preds, all_labels = [], []
with torch.no_grad():
    for batch in tqdm(test_dataloader, desc="Test Evaluation"):
        input_ids = batch['input_ids'].to('cuda')
        attention_mask = batch['attention_mask'].to('cuda')
        labels = batch['labels'].to('cuda')

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        loss = loss_fn(outputs, labels)
        test_loss += loss.item()
        test_acc += calc_accuracy(outputs, labels)

        all_preds.extend(torch.argmax(outputs, axis=1).cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        # Memory Optimization
        del outputs, input_ids, attention_mask, labels
        torch.cuda.empty_cache()

test_loss /= len(test_dataloader)
test_acc /= len(test_dataloader)
test_f1 = f1_score(all_labels, all_preds, average='weighted')
print(f"[Test Results] Loss: {test_loss:.4f}, Accuracy: {test_acc:.4f}, F1-Score: {test_f1:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=['Positive', 'Neutral', 'Negative']))
```

```python
# 데이터 전처리 함수 추가
def preprocess_comments(df):
    # 결측치 처리
    df['comment'] = df['comment'].fillna("").astype(str)
    return df

# 수정된 analyze_comments_and_save 함수
def analyze_comments_and_save(final_df, model, batch_size=64, max_len=128):
    # 데이터 전처리
    final_df = preprocess_comments(final_df)  # 추가된 전처리

    # 모델 평가 모드 설정
    model.eval()

    # 데이터셋 생성
    dataset = BERTDataset(final_df, tokenizer, max_len)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    # 예측
    predictions = []
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to('cuda')
            attention_mask = batch['attention_mask'].to('cuda')
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            batch_predictions = torch.argmax(outputs, dim=1).cpu().numpy()
            predictions.extend(batch_predictions)

    # 감정 재분류
    reclassified_emotions = reclassify_emotions(predictions)

    # 결과 추가
    final_df['Feelings'] = reclassified_emotions

    # comments.csv 파일 저장
    comments_file = os.path.join(SAVE_DIR, "comments.csv")
    final_df.to_csv(comments_file, index=False, encoding='utf-8-sig')
    print(f"comments.csv 파일 생성 완료: {comments_file}")

    # 감정 비율 계산 및 저장
    ratio_counts = final_df['Feelings'].value_counts()
    ratio_df = pd.DataFrame({'Emotion': ['Positive', 'Neutral', 'Negative'],
                             'Count': [ratio_counts.get('Positive', 0), ratio_counts.get('Neutral', 0), ratio_counts.get('Negative', 0)]})
    ratio_file = os.path.join(SAVE_DIR, "ratio.csv")
    ratio_df.to_csv(ratio_file, index=False, encoding='utf-8-sig')
    print(f"ratio.csv 파일 생성 완료: {ratio_file}")

    # 분기별 감정 계산
    final_df['Date'] = pd.to_datetime(final_df['published_at'])  # 날짜 변환
    min_date, max_date = final_df['Date'].min(), final_df['Date'].max()
    date_ranges = pd.date_range(min_date, max_date, periods=5)  # Bin 경계 생성

    # 'quarters'의 길이를 date_ranges보다 하나 적게 맞추기
    quarters = ['Q1', 'Q2', 'Q3', 'Q4']

    # pd.cut 사용 시 경계와 레이블 길이 맞춤
    final_df['Quarter'] = pd.cut(final_df['Date'], bins=date_ranges, labels=quarters, right=False)

    # 분기별 감정 개수 계산
    quarter_counts = final_df.groupby(['Quarter', 'Feelings']).size().unstack(fill_value=0)
    quarter_counts = quarter_counts.reindex(columns=['Positive', 'Neutral', 'Negative'], fill_value=0)
    quarter_counts.insert(0, 'Date', [d.strftime('%Y-%m-%d') for d in date_ranges[:-1]])  # 분기 시작 날짜 추가

    count_file = os.path.join(SAVE_DIR, "count.csv")
    quarter_counts.reset_index(drop=True, inplace=True)
    quarter_counts.to_csv(count_file, index=False, encoding='utf-8-sig')
    print(f"count.csv 파일 생성 완료: {count_file}")

    return comments_file, ratio_file, count_file

# 실행 코드
if __name__ == "__main__":
    # 입력 데이터셋 로드
    input_file_path = "/content/drive/MyDrive/캡스톤/comments.csv"
    final_df = pd.read_csv(input_file_path)

    # 모델 초기화
    class BERTClassifier(torch.nn.Module):
        def __init__(self, bert, dr_rate=0.5):
            super(BERTClassifier, self).__init__()
            self.bert = bert
            self.dropout = torch.nn.Dropout(p=dr_rate)
            self.classifier = torch.nn.Linear(self.bert.config.hidden_size, 3)

        def forward(self, input_ids, attention_mask):
            output = self.bert(input_ids=input_ids, attention_mask=attention_mask)
            pooled_output = output.pooler_output
            dropped_output = self.dropout(pooled_output)
            return self.classifier(dropped_output)

    model = BERTClassifier(bertmodel).to('cuda')

    # 모델 로드
    model.load_state_dict(torch.load("/content/best_kobert_model_epoch_12.pt"))
    model.to('cuda')

    # 감정 분석 실행
    comments, ratios, counts = analyze_comments_and_save(final_df, model)
    print(f"모든 파일 생성 완료:\n- Comments: {comments}\n- Ratios: {ratios}\n- Counts: {counts}")
```

```python
import pandas as pd
import re
from datetime import datetime
import os
from torch.utils.data import DataLoader
import torch
import numpy as np
from transformers import BertTokenizer
from torch.nn.functional import softmax

# 저장 디렉토리 설정
SAVE_DIR = "/content/drive/MyDrive/캡스톤/"
os.makedirs(SAVE_DIR, exist_ok=True)

# KoBERT 모델 로드
from transformers import BertModel
from transformers import BertForSequenceClassification

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_path = "best_kobert_model.pt"
tokenizer = BertTokenizer.from_pretrained('skt/kobert-base-v1')
model = BertForSequenceClassification.from_pretrained('skt/kobert-base-v1', num_labels=3)  # 3개의 감정
model.load_state_dict(torch.load(model_path))
model.to(device)
model.eval()

# 텍스트 전처리 함수
def clean_text(text):
    text = re.sub(r"[^\uAC00-\uD7A3a-zA-Z0-9\s!?.,]", "", text)  # 한글, 영어, 숫자, 특수문자만 남김
    text = re.sub(r"\s+", " ", text)  # 연속된 공백 제거
    return text.strip()

# 감정 분석 함수
def analyze_emotions_with_reclassification(file_path):
    # CSV 파일 읽기
    df = pd.read_csv(file_path)

    # 날짜 열 처리
    if 'Date' not in df.columns:
        if 'Published At' in df.columns:
            df['Date'] = pd.to_datetime(df['Published At'])  # `Published At` → `Date`
        else:
            raise KeyError("입력 데이터에 'Date' 또는 'Published At' 열이 없습니다. 날짜 정보를 추가해주세요.")

    # 전처리: 특수 기호 및 공백 처리
    df['Cleaned_Comment'] = df['Comment'].fillna('').apply(clean_text)
    df = df.dropna(subset=['Cleaned_Comment'])  # 전처리 결과 None인 행 제거

    # 모델 기반 감정 분석
    sentences = df['Cleaned_Comment'].tolist()
    predictions = []

    # 배치 크기 설정
    batch_size = 32
    for i in range(0, len(sentences), batch_size):
        batch_sentences = sentences[i:i + batch_size]
        inputs = tokenizer(batch_sentences, padding=True, truncation=True, return_tensors="pt", max_length=128)
        inputs = {key: val.to(device) for key, val in inputs.items()}

        with torch.no_grad():
            outputs = model(**inputs)
            probs = softmax(outputs.logits, dim=1)
            preds = torch.argmax(probs, axis=1).cpu().numpy()
            predictions.extend(preds)

    # 기존 감정을 재분류
    reclassified_emotions = ["Positive" if pred == 0 else "Neutral" if pred == 1 else "Negative" for pred in predictions]

    # 결과 추가
    df['Feelings'] = reclassified_emotions

    # comments.csv 파일 저장
    comments_file = os.path.join(SAVE_DIR, "comments.csv")
    df.to_csv(comments_file, index=False, encoding='utf-8-sig')

    # 감정 비율 계산 및 저장
    ratio_counts = df['Feelings'].value_counts()
    ratio_df = pd.DataFrame({'Emotion': ['Positive', 'Neutral', 'Negative'],
                             'Count': [ratio_counts.get('Positive', 0), ratio_counts.get('Neutral', 0), ratio_counts.get('Negative', 0)]})
    ratio_file = os.path.join(SAVE_DIR, "ratio.csv")
    ratio_df.to_csv(ratio_file, index=False, encoding='utf-8-sig')

    # 분기별 감정 계산
    min_date, max_date = df['Date'].min(), df['Date'].max()
    date_ranges = pd.date_range(min_date, max_date, periods=5)
    quarters = ['Q1', 'Q2', 'Q3', 'Q4']
    df['Quarter'] = pd.cut(df['Date'], bins=date_ranges, labels=quarters[:-1], right=False)

    quarter_counts = df.groupby(['Quarter', 'Feelings']).size().unstack(fill_value=0)
    quarter_counts = quarter_counts.reindex(columns=['Positive', 'Neutral', 'Negative'], fill_value=0)
    quarter_counts.insert(0, 'Date', [d.strftime('%Y-%m-%d') for d in date_ranges[:-1]])

    count_file = os.path.join(SAVE_DIR, "count.csv")
    quarter_counts.reset_index(drop=True, inplace=True)
    quarter_counts.to_csv(count_file, index=False, encoding='utf-8-sig')

    return comments_file, count_file, ratio_file

# 실행 코드
if __name__ == "__main__":
    input_file_path = "/content/drive/MyDrive/캡스톤/youtube_comments.csv"  # 입력 파일 경로
    if os.path.exists(input_file_path):
        comments, counts, ratios = analyze_emotions_with_reclassification(input_file_path)
        print(f"모든 파일이 성공적으로 생성되었습니다.\n- Comments: {comments}\n- Counts: {counts}\n- Ratios: {ratios}")
    else:
        print(f"입력 파일이 존재하지 않습니다: {input_file_path}")
```
